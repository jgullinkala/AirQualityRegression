```python
#checking for duplicates
air_quality_df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns


plt.figure(figsize = (20,20))

for i, column in enumerate(air_quality_df.columns,1):
    plt.subplot(5,3,i)
    sns.boxplot(data=air_quality_df, x=column)
    plt.xlabel(column)

plt.show()

# We can see some outliers - It looks like -200, 
```

```python
for col in air_quality_df.columns:
    print(col, ":" , len(air_quality_df[air_quality_df[col]== -200]), 'Number of times -200 appears in the column')
    print(col, ":" , len(air_quality_df[air_quality_df[col]== -200])/air_quality_df[col].count()*100, 'Number of times -200 appears in the column')
```

```python
# Review Categorical columns

air_quality_df.select_dtypes("object").head()
```

```python
# Data cleaning and validations

air_quality_df['Date'].min(), air_quality_df['Date'].max() # Format is not correct as min is showng 2005 data and max is showing 2004

```

```python
# Fixing the data formatting issues

air_quality_df['Date'] = pd.to_datetime(air_quality_df['Date'], format = '%m/%d/%Y')

#checking the min and max
air_quality_df['Date'].min(), air_quality_df['Date'].max()  # corrected formats

```
