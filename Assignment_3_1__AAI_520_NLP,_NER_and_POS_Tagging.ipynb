```python
# Apply NER techniques to identify named entities (such as persons, organizations, locations, etc.) within the text).

import spacy
nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
  doc = nlp(text)
  entities = [(ent.text, ent.label_) for ent in doc.ents]
  return entities

# Apply the extract_entities function to the 'preprocessed_text' column and store the results in the 'entities' column
data["entities"] = data["preprocessed_text"].apply(lambda x: extract_entities(" ".join(x))) # Join the list of words back into a string
```

# Introduction

This report summarizes the process of applying Named Entity Recognition (NER) and Part-of-Speech (POS) tagging techniques to the Climate Fever dataset. The goal was to identify key entities and their grammatical roles within the claims related to climate change.


# Approach

1. **Data Loading and Preprocessing:**
   - The Climate Fever dataset was loaded into a Pandas DataFrame.
   - The 'claim' column, containing the text of the claims, was preprocessed:
     - Tokenization: Splitting the text into individual words.
     - Stemming: Reducing words to their root form.
     - Lemmatization: Converting words to their base or dictionary form.
     - Stop word removal: Eliminating common words (e.g., "the," "a," "is") that don't carry significant meaning.

2. **Named Entity Recognition (NER):**
   - The spaCy library was used to perform NER.
   - The `extract_entities` function was applied to the preprocessed text to identify named entities (e.g., persons, organizations, locations, dates) and their labels.

3. **Part-of-Speech (POS) Tagging:**
   - The `extract_pos_tags` function was used to assign POS tags to each word in the claims.
   - This helps understand the grammatical role of each word within the sentence.

4. **Analysis and Visualization:**
   - The identified entities and their corresponding POS tags were printed for each claim.
   - spaCy's `displacy` was used to visualize the entities and dependency parsing within the claims.


# Findings

- **Key Entities:** The NER process successfully identified various entities within the claims, including:
    - **Organizations:** (e.g., "IPCC," "NASA")
    - **Locations:** (e.g., "Arctic," "United States")
    - **Dates:** (e.g., "2023")
    - **Persons:** (e.g., "Trump")
- **POS Tagging Insights:** POS tagging provided insights into the grammatical structure of the claims, helping to understand how entities are used within the context of the sentences.
- **Visualization:** The visualizations using `displacy` allowed for a better understanding of the relationships between words and entities within the claims.





# Challenges

- **Ambiguity in Text:** Some claims contained ambiguous language, making it difficult for the NER model to accurately identify entities.
- **Contextual Understanding:** NER and POS tagging are limited in their ability to understand the full context of the claims. For example, they may not always correctly identify the meaning of a word based on its surrounding words.
- **Data Quality:** The quality of the data could affect the accuracy of the NER and POS tagging results.


# Conclusion

This analysis demonstrates the effectiveness of NER and POS tagging in extracting valuable information from the Climate Fever dataset. The identified entities and their grammatical roles can be further analyzed to gain deeper insights into the claims related to climate change.

Future work could focus on improving the accuracy of NER and POS tagging by using more advanced techniques, such as incorporating contextual information and training custom models for climate-related text.
